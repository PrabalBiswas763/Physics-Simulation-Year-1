```python
#Simulating a simple PENDULUM
import numpy as np
import math
import matplotlib.pyplot as plt
def simple_pendulum(time,theta):

  g=9.8
  L=1
  t=0.0000
  dt=0.01
  omega=0
  time_list=[]
  theta_list=[]
  steps=int(time/dt)
  for i in range(steps):

    time_list.append(t)
    alpha=-(g/L)*math.sin(theta)
    theta_list.append(theta)
    omega+=alpha*dt
    theta+=omega*dt
    t+=dt

  return time_list,theta_list
#print(simple_pendulum(10,0.3))

def plot_graph():
   x,y=simple_pendulum(10,0.3)
   plt.title('Simple Pendulum')
   plt.plot(x,y)
   plt.xlabel('time')
   plt.ylabel('theta')
   plt.show()
plot_graph()
```

```python
#Pendulum with Friction in consideration
import numpy as np
import math
import matplotlib.pyplot as plt
def pendulum_with_friction(time,theta,g,l,gamma,m):
   time_l=[]
   theta_l=[]
   ke=0
   pe=m*g*l*(1-math.cos(theta))
   te_l=[]
   omega_l=[]
   t=0
   dt=0.001
   omega=0
   while t< time:
    time_l.append(t)
    theta_l.append(theta)
    te_l.append(ke+pe)
    omega_l.append(omega)
    alpha=-(g/l)*math.sin(theta)-gamma*omega
    omega+=alpha*dt
    theta+=omega*dt
    ke=0.5*m*l*l*omega*omega
    pe=m*g*l*(1-math.cos(theta))
    t+=dt
   return time_l,theta_l,te_l,omega_l

def graph_plot():
  x,y,c,d=pendulum_with_friction(50,0.3,9.8,1,0.1,1)
  plt.plot(x,y, 'r-',)
  plt.xlabel('time')
  plt.ylabel('theta')
  plt.title('Pendulum with Friction')
  plt.show()
  plt.plot(x,c, 'y-')
  plt.xlabel('time')
  plt.ylabel('total energy')
  plt.title('Total Energy')
  plt.show()
  plt.plot(y,d,'g-')
  plt.xlabel('theta')
  plt.ylabel('omega')
  plt.title('Angular Velocity')
  plt.show()
graph_plot()
```
